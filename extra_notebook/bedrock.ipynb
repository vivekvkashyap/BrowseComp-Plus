```python
import boto3
import json
from botocore.exceptions import ClientError
```

```python
# Import necessary libraries
import os
from google.colab import userdata

os.environ["AWS_BEARER_TOKEN_BEDROCK"] = "your-key-here"
```

```python
# Create a Bedrock Runtime client in the AWS Region where the model is available (e.g., us-east-1).
# Ensure your AWS credentials are set up (e.g., via environment variables or AWS CLI).
client = boto3.client("bedrock-runtime", region_name="us-east-1")

# Set the model ID for Anthropic Claude 3 Haiku (text generation model).
model_id = "anthropic.claude-3-haiku-20240307-v1:0"

# Define the prompt for code generation.
prompt = """Write a Python function that takes a string as input and returns it reversed.
Include comments explaining the code. Do not use built-in reverse methods like [::-1]."""

# Format the request payload using Claude's messages API.
native_request = {
    "anthropic_version": "bedrock-2023-05-31",
    "max_tokens": 512,  # Adjust based on expected output length.
    "temperature": 0.7,  # Controls randomness (0.0-1.0); lower for more deterministic output.
    "messages": [
        {
            "role": "user",
            "content": prompt
        }
    ]
}

# Convert the native request to JSON.
request_body = json.dumps(native_request)

try:
    # Invoke the model.
    response = client.invoke_model(
        modelId=model_id,
        body=request_body,
        contentType="application/json",
        accept="application/json"
    )

    # Decode the response body.
    model_response = json.loads(response["body"].read())

    # Extract the generated text (for Claude, it's in content[0].text).
    generated_code = model_response["content"][0]["text"]
    print("Generated Code:\n")
    print(generated_code)

except (ClientError, Exception) as e:
    print(f"ERROR: Can't invoke '{model_id}'. Reason: {e}")
    exit(1)
```
